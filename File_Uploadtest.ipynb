```python
# ファイルアップロードテストプログラム
from google.colab import files

def upload_and_read_file():
    print("Google colab上でテキストファイルをアップロードし、表示するプログラムです。\n読み込みに成功するとテキストの内容が出力されます。\n")
    print("サイズの大きいファイルは正常にアップロード処理ができない場合がありますのでご注意ください。\n")
    print("アップロードされたファイルはGoogle Colabのセッション内で一時的に保存されます。\nこのデータはセッション終了後に削除され、Googleのサーバーに長期的に保存されることはありません。\n")
    print("プライバシーを気にする場合、個人情報を含むファイルはアップロードしないことを推奨します。\n")
    print("テキストファイルをアップロードしてください(テキストファイル以外はアップロードできません。):")
    uploaded = files.upload()  # ファイルをアップロード
    for filename in uploaded.keys():
        print(f"\nアップロードされたファイル名: {filename}")
        print("\n内容:\n")
        print(uploaded[filename].decode("utf-8"))  # ファイルの内容を表示

# 実行
upload_and_read_file()
```
