```python
#LINE-Log-Manager
#Copyright (c) Payao The 84th/Meta Kernel Studios,2025. All rights reserved.
#This code was partially generated with assistance from OpenAI's ChatGPT.
# 必要なライブラリのインポート
from google.colab import files
from datetime import datetime, timedelta
import re

# グローバル変数
lines = []

def load_log_file():
    """ログファイルをアップロードして読み込む関数"""
    global lines
    print("テキストファイルをアップロードしてください(ファイルサイズが大きい場合、処理に時間がかかる場合があります):")
    uploaded = files.upload()  # ファイルアップロード
    for filename, content in uploaded.items():
        lines = content.decode("utf-8").splitlines()  # ファイル内容をリストに保存
        print(f"ファイル '{filename}' が正常に読み込まれました。")

def search_logs():
    """ログを検索し、結果を出力する関数"""
    if not lines:
        print("ログファイルがロードされていません。最初にログファイルをアップロードしてください。")
        return

    search_type = input("検索タイプを選択してください（'ワード' または '日付'）: ").strip()
    results = []

    if search_type == "ワード":
        search_word = input("検索ワードを入力してください: ").strip()
        if search_word:
            results = [line for line in lines if search_word in line]
    elif search_type == "日付":
        try:
            year = int(input("検索する年を入力してください（例: 2024）: "))
            month = int(input("検索する月を入力してください（例: 1）: "))
            day = int(input("検索する日を入力してください（例: 1）: "))
            start_date = datetime(year, month, day)
            end_date = start_date + timedelta(days=1)
            results = extract_date_range_logs(start_date, end_date)
        except ValueError:
            print("無効な日付が入力されました。")
            return
    else:
        print("無効な検索タイプです。")
        return

    print(f"--- 検索結果 ({len(results)} 件がヒットしました) ---")
    if results:
        for line in results:
            print(line)
    else:
        print("一致する結果が見つかりませんでした。")

    export_option = input("検索結果をエクスポートしますか？（y/n）: ").strip().lower()
    if export_option == 'y':
        export_results(results)

def extract_date_range_logs(start_date, end_date):
    """指定された日付範囲のログを抽出する"""
    result = []
    current_date = None

    for line in lines:
        # 日付を含む行をチェック
        date_match = re.match(r"(\d{4}/\d{1,2}/\d{1,2})", line)
        if date_match:
            try:
                current_date = datetime.strptime(date_match.group(1), "%Y/%m/%d")
            except ValueError:
                continue

        if current_date and start_date <= current_date < end_date:
            result.append(line)

    return result

def export_results(results):
    """検索結果をダウンロードできるようにエクスポートする関数"""
    execution_date = datetime.now().strftime("%Y/%m/%d %H:%M:%S")
    header = (
        f"LINEログマネージャー\n"
        f"Copyright (c) Payao The 84th/Meta Kernel Studios,2025. All rights reserved.\n"
        f"This code was partially generated with assistance from OpenAI's ChatGPT.\n"
        f"実行日: {execution_date}\n"
        f"--- 検索結果 ({len(results)} 件がヒットしました) ---\n"
    )

    result_text = header + "\n".join(results)
    with open("search_results.txt", "w", encoding="utf-8") as f:
        f.write(result_text)

    print("検索結果が 'search_results.txt' にエクスポートされました。")
    print("ダウンロード後は管理簡易化のためファイル名を変更する事を推奨します。")
    files.download("search_results.txt")

# メインの処理
def main():
    while True:
        print("LINEログマネージャー")
        print("Copyright (c) Payao The 84th/Meta Kernel Studios,2025. All rights reserved.")
        print("This code was partially generated with assistance from OpenAI's ChatGPT.")
        print("\nメニュー:")
        print("1. ログファイルをアップロード")
        print("2. 検索を実行")
        print("3. 終了")
        print("現在、ダウンローダーの仕様により一度プログラムを終了しないとエクスポートのダウンロードが開始されません。")
        print("[3]を選択し一度終了するとダウンロードが開始されます。")

        choice = input("選択してください（1/2/3）: ").strip()
        if choice == "1":
            load_log_file()
        elif choice == "2":
            search_logs()
        elif choice == "3":
            print("プログラムを終了します。")
            break
        else:
            print("無効な選択です。再度選択してください。")

# プログラム開始
main()
```
